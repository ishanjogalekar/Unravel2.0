```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('student_prediction.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.fillna(df.mean())
```

```python
df.isnull().values.any()
```

```python
df.columns 
```

```python
df.nunique()
```

```python
df.drop_duplicates(inplace=True,keep=False)
```

```python
correlation1 = df.corr()
print(correlation1)
```

```python
sns.set(rc={'figure.figsize':(30,30)})
sns.heatmap(correlation1, annot=True,linewidths=3, xticklabels=correlation1.columns, 
 yticklabels=correlation1.columns, linecolor='white', cbar=True)
```

```python
grade = df['GRADE']
salary = df['SALARY']
pt = df['KIDS']
hs = df['HS_TYPE']
fj = df['FATHER_JOB']
age = df['AGE']

data = pd.DataFrame(zip(grade,salary,pt,hs,fj,age))
data.columns = ['Grades','Salary','Parental_Status','HS_Type','Father_Job','Age']
```

```python
data.shape
```

```python
data.describe()
```

```python
data.info()
```

```python
correlation = data.corr()
print(correlation)
```

### Correlation HeatMap

```python
sns.set(rc={'figure.figsize':(13,13)})
sns.heatmap(correlation, annot=True,linewidths=.5, xticklabels=correlation.columns, 
 yticklabels=correlation.columns, linecolor='white', cbar=True)
```

# Plots

### Barplot

```python
# Barplot
labels=['Fail','DD','DC','CC','CB','BB','BA','AA']
plt.figure(figsize=(13,13))
sns.set_theme(style='whitegrid')
plt.subplot(2,2,1)
sns.barplot(data=data,y='Salary',x='Grades').set(title = "Salary Vs. Grade")
plt.subplot(2,2,2)
sns.barplot(data=data,y='Parental_Status',x='Grades').set(title = "Parental-Status Vs. Grade")
plt.subplot(2,2,3)
sns.barplot(data=data,y='HS_Type',x='Grades').set(title = "High-School Status Vs. Grade")
plt.subplot(2,2,4)
sns.barplot(data=data,y='Father_Job',x='Grades').set(title = "Father Job Vs. Grade")
```

### Violin Plot

```python
# violin plot
plt.figure(figsize=(13,13))
sns.set_theme(style='whitegrid')
plt.subplot(2,2,1)
sns.violinplot(data=data,y='Salary',x='Grades').set(title = "Salary Vs. Grade")
plt.subplot(2,2,2)
sns.violinplot(data=data,y='Parental_Status',x='Grades').set(title = "Parental-Status Vs. Grade")
plt.subplot(2,2,3)
sns.violinplot(data=data,y='HS_Type',x='Grades').set(title = "High-School Status Vs. Grade")
plt.subplot(2,2,4)
sns.violinplot(data=data,y='Father_Job',x='Grades').set(title = "Father Job Vs. Grade")
```

### Line Plot

```python
# line plot
plt.figure(figsize=(13,13))
sns.set_theme(style='whitegrid')
plt.subplot(2,2,1)
sns.lineplot(data=data,y='Salary',x='Grades',linewidth=2.5).set(title = "Salary Vs. Grade")
plt.subplot(2,2,2)
sns.lineplot(data=data,y='Parental_Status',x='Grades',color='orange',linewidth=2.5).set(title = "Parental-Status Vs. Grade")
plt.subplot(2,2,3)
sns.lineplot(data=data,y='HS_Type',x='Grades',color='red',linewidth=2.5).set(title = "High-School Status Vs. Grade")
plt.subplot(2,2,4)
sns.lineplot(data=data,y='Father_Job',x='Grades',color='green',linewidth=2.5).set(title = "Father Job Vs. Grade")
```

### Density and Area Plots

```python
data.plot(kind='density', subplots=True, layout=(3,3), sharex=False , figsize =(15,15))
plt.show()
```

```python
data.plot(kind='area', subplots=True, layout=(3,3), sharex=False , figsize =(15,15))
plt.show()
```

```python
plt.figure(figsize=(30, 30))
sns.pairplot(data,hue='Grades',height=3, diag_kind='hist')
```

```python
data = data.sort_values(by=['Age','Grades'],ascending=[True,True])
print(data)
data.describe()
```

### Grade Vs Age Plot

```python
plt.figure(figsize=(15, 15))
plt.ylim(0, 7)
sns.boxplot(data=data,y='Grades',x='Age').set(title = "Grade Vs. Age")
```
